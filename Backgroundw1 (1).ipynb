The Battle of Neighborhoods

Introduction

1.1 Background


I have been living in the City of Youngstown for almost two years now. In my free time, I enjoy jogging around campus and downtown. I love this city because I had so much memories here, but eventually I realized there are things I do not enjoy. For this Capstone, I decided to work on solving one of the problem that the people here have been facing for a long time, in the hope that there would be some improvement.

In 2020, the City of Youngstown has officially become the poorest city in the United States. Major industries moving out of the area since the 70s is the main reason for the decline of this city. The closure of GM-Lordstown in 2018 also has such a great impact on the area that the city from being the second-poorest in the country - now rised to the first position.

As more and more young citizens of the city move away to find better opportunity, the condition of the city, including facilities and safety, get worse. The city is considered "food-desert", as there is so little access to groceries for people in the area. 

1.2 Problem

Access to groceries is neccesary for people anywhere. Lacking this commoditty is a factor that drive people away from the area. This project will analyse the locations of current groceries store in the area, identify neighbourhoods without access to groceries, and potential places to open local groceries store based on certain criteria.
